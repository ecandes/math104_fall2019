# Solution to computational problem

The code below generate 5 vectors in 100-dimensional space with random orientation arranged in a matrix $X$. We then make a $5 \times 50$ matrix $B$ with random entries and compute 

$$
A = X B
$$

The columns of $A$ are linear combinations of those of $X$ and therefore live in a subspace of dimension 5. The rank of $A$ is obviously equal to 5.

```python
import numpy as np

m, n = 100, 50
X = np.random.randn(m,5)
B = np.random.randn(5,n)
A = X@B
```

We now take the SVD of $A$ and plot the singular values.

```python
U, S, V = np.linalg.svd(A)  
V = V.T

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use("seaborn")
plt.rcParams["figure.figsize"] = (8, 4)
sns.set_context("talk", font_scale=1)

plt.plot(S,'*')
plt.title('Singular values of A')
plt.show()
```

A plot on a log scale may be more informative

```python
plt.semilogy(S,'*')
plt.title('Singular values of A')
plt.show()
```

We now perturb the columns of the matrix by adding a bit of noise to each entry of $A$. This makes the matrix have full column rank; that is, the rank is now 50. 

```python
perturbedA = A + 1e-2*np.random.randn(m,n)
```

We then recompute the singular values and plot them. 

```python
_ , perturbedS, _ = np.linalg.svd(perturbedA)

plt.rcParams["figure.figsize"] = (8, 4)
sns.set_context("talk", font_scale=1)

plt.semilogy(perturbedS,'*')
plt.title('Singular values of perturbed matrix')
plt.show()
```

We see that all the singular values are now nonzero conforming that the rank is 50. However, we see that 45 singular values are quite small indicating that columns of $A$ lie near a subspace of dimension 5.
